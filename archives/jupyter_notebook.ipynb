```python
#Load pdf document
loader = PyPDFLoader("pages-29-deed-sample.pdf")
pages = loader.load_and_split()
print(pages[0])
```

```python
#chunking the document
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

docs = text_splitter.split_documents(pages)
print(type(pages))
print(type(pages[0]))

print(docs[0])
```

```python
#load embeddings
embeddings = OpenAIEmbeddings()

library = FAISS.from_documents(docs, embeddings)
```

```python
#retrieve answer
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)
query_1 = "What is the purpose of this document?"

context = library.similarity_search(query_1, k=3)
# print(f"Answer to query 1: {query_1_answer[0].page_content}")

# retriever = library.as_retriever()

prompt = f"""You are a legal assistant helping summarize legal documents. 
Using the context below, answer the question concisely and accurately.
Context: {context}
Question: {query_1}"""


response = llm.invoke(prompt)
print(f"Answer to query 1: {response.content}")
```
